```python
# use different figure size
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(5, 3))  
plt.plot(x, y)
plt.title("Small Figure (5x3)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))  
plt.plot(x, y)
plt.title("Medium Figure (8x5)")
plt.grid(True)
plt.show()

plt.figure(figsize=(12, 6))  
plt.plot(x, y)
plt.title("Large Figure (12x6)")
plt.grid(True)
plt.show()
```

```python
# Annotation
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, marker='o')

plt.annotate('This is 30', (3, 30))

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Annotation Example')

plt.grid(True)
plt.show()
```

```python
# change axis and background color
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.xlabel('Class')
plt.ylabel('Student')


plt.gcf().set_facecolor('lightyellow')  


plt.gca().set_facecolor('lavender')     


plt.gca().spines['bottom'].set_color('red')
plt.gca().spines['left'].set_color('blue')


plt.scatter(x, y, color='green', marker='*')

plt.title('Colored Background & Axis')
plt.grid(True)

plt.show()
```

```python
# twinx and twiny
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [50, 40, 30, 20, 10]

fig, ax1 = plt.subplots()


ax1.plot(x, y1, color='blue')
ax1.set_xlabel('Bottom X-Axis')
ax1.set_ylabel('Left Y-Axis')


ax2 = ax1.twinx()
ax2.plot(x, y2, color='red')
ax2.set_ylabel('Right Y-Axis')


ax3 = ax1.twiny()
ax3.set_xlabel('Top X-Axis')

plt.title("Twin X and Twin Y Axis")
plt.show()
```
